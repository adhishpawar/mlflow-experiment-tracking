```python
import mlflow
import numpy as np
import json
```

```python
mlflow.set_experiment("CustomExperiment")
```

```python
with mlflow.start_run():
    # Log parameters
    mlflow.log_param("learning_rate", 0.01)
    mlflow.log_param("epochs", 20)

    # Generate fake metric values
    for step in range(1, 6):
        mlflow.log_metric("accuracy", np.random.rand(), step=step)

    # Create a custom artifact (json)
    artifact_data = {"message": "MLflow works!", "version": 1}
    with open("meta.json", "w") as f:
        json.dump(artifact_data, f)

    mlflow.log_artifact("meta.json")

    print("Run logged with custom artifacts.")
```
