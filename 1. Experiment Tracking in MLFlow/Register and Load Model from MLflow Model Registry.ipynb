```python
# Notebook 3: Model Registry â€“ Save, Register & Load

import mlflow
import mlflow.sklearn
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
```

```python
# Load data
data = load_diabetes()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=42
)
```

```python
# Set experiment
mlflow.set_experiment("RegistryExperiment")
```

```python
with mlflow.start_run() as run:
    model = RandomForestRegressor(n_estimators=50)
    model.fit(X_train, y_train)

    # Log with tracking
    mlflow.sklearn.log_model(model, "rf_model")

    # Register model
    model_uri = "runs:/{}/rf_model".format(run.info.run_id)
    model_name = "DiabetesRFModel"

    print("Registering model...")
    mlflow.register_model(model_uri, model_name)

print("Model registered.")
```

```python
# Load the latest version from registry
loaded_model = mlflow.pyfunc.load_model(f"models:/{model_name}/latest")
print("Model loaded from registry!")
```
